```python
import numpy as np
import matplotlib.pyplot as plt


# Diese vier Parameter sollen Sie verändern
rho_fluid = 1000; # Dichte des Porenfluids, typischerweise Wasser [kg/m^3]
rho_matrix = 2500; # Dichte der Gesteinsmatrix [kg/m^3]
P=0.25; # Anteil des Porenraums im Gestein, P = 0.2 würde zum Beispiel 20% Porosität bedeuten
T=0.5; # Anteil der Poren, die mit dem Porenfluid gefüllt sind. Ein T = 0.5 würde heißen, dass die Hälfte
# der Poren mit dem Fluid gefüllt sind, der Rest ist mit Luft (mit der Dichte = kg/^3) gefüllt 



# Ab hier nichts mehr am Skript verändern
v=np.linspace(0,1,11);
rho1=2500*(1-P)+1000.*P*v;
Rho2=2500*(1-v)+1000*v*T;
Rho3=2500*(1-v)+1000*v*v;

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.plot(v,rho1,'.-')
plt.ylabel('Dichte [kg/m^3]')
plt.xlabel('Wasseranteil')
plt.title('Dichte eines Gesteins mit einer Matrixdichte von %d kg/^3 und einer Porosität \n von %d Prozent als Funktion der Wassersättigung'%(rho_matrix,P*100))

plt.subplot(2,2,2);
plt.plot(v,Rho2,'.-')
plt.title('Dichte eines Gesteins  mit einer Matrixdichte von %d kg/^3 und einer Sättigung \n von %d Prozent als Funktion der Porosität'%(rho_matrix,T*100))
plt.ylabel('Dichte [kg/m^3]')
plt.xlabel('Porosität')

plt.subplot(2,2,3); 
plt.plot(v,Rho3,'.-')
plt.title('Dichte eines Gesteins  mit einer Matrixdichte von %d kg/^3 und als Funktion der \n Sättigung und Porosität \n (z.B. bei 20 Prozent Porosität sind 20 Prozent des POrenraums gesättigt)'%(rho_matrix))
plt.xlabel('P,Theta')
plt.ylabel('Dichte [kg/m^3]')
plt.tight_layout()
plt.show()
```
