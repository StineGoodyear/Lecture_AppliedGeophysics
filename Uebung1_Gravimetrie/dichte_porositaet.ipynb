# Jupyter Notebook zur Dichtebestimmung poröser Gesteine
In der zweiten Zelle dieses Notebooks (diese Textzelle ist die erste Zelle) werden einer Gesteinsmatrix (rho_matrix) und einem Fluid Dichtewerte und Werte für den Porenraumanteil (P) und den Fluidanteil (T) zugeordnet.
Diese Werte können Sie beliebig ändern.

In der dritten Zelle wird daraus die Gesamtdichte des porösen Gesteins berechnet und Abbildungen erzeugt, die die Abhängigkeiten zwischen den jeweiligen Dichten sowie dem Porenraum- und dem Fluidanteil darstellen. Nehmen Sie in dieser Zelle bitte keine Änderungen vor.

Die Programme in den jeweiligen Zellen werden ausgeführt, indem Sie in der Zelle Strg+Enter drücken. Oder in der Menüleiste oben den "Run"-Button.


```python
import numpy as np
import matplotlib.pyplot as plt

# Diese vier Parameter können Sie verändern
rho_fluid = 1000; # Dichte des Porenfluids, typischerweise Wasser [kg/m^3]
rho_matrix = 2500; # Dichte der Gesteinsmatrix [kg/m^3]
P=0.25; # Anteil des Porenraums im Gestein, P = 0.25 würde zum Beispiel 25% Porosität bedeuten
T=0.5; # Anteil der Poren, die mit dem Porenfluid gefüllt sind. Ein T = 0.5 würde heißen, dass die Hälfte
# der Poren mit dem Fluid gefüllt sind, der Rest ist mit Luft (mit der Dichte = kg/m^3) gefüllt 
```

```python
# Ab hier nichts mehr am Skript verändern
v=np.linspace(0,1,11);
rho1=rho_matrix*(1-P)+rho_fluid*P*v;
rho2=rho_matrix*(1-v)+rho_fluid*T*v;
#Rho3=rho_matrix*(1-v)+rho_fluid*v*v;

plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
plt.plot(v,rho1,'.-')
plt.ylabel('Dichte [kg/m^3]')
plt.xlabel('Wasseranteil')
plt.title('Dichte eines Gesteins mit einer Matrixdichte von %d kg/m^3 und einer Porosität \n von %d Prozent als Funktion der Wassersättigung'%(rho_matrix,P*100))

plt.subplot(2,2,3);
plt.plot(v,rho2,'.-')
plt.title('Dichte eines Gesteins  mit einer Matrixdichte von %d kg/m^3 und einer Sättigung \n von %d Prozent als Funktion der Porosität'%(rho_matrix,T*100))
plt.ylabel('Dichte [kg/m^3]')
plt.xlabel('Porosität')

#plt.subplot(3,2,5); 
#plt.plot(v,Rho3,'.-')
#plt.title('Dichte eines Gesteins  mit einer Matrixdichte von %d kg/^3 und als Funktion der \n Sättigung und Porosität \n (z.B. bei 20 Prozent Porosität sind 20 Prozent des Porenraums gesättigt)'%(rho_matrix))
#plt.xlabel('P,T')
#plt.ylabel('Dichte [kg/m^3]')
#plt.tight_layout()
#plt.show()
```
